```python
import pandas as pd

import matplotlib.pyplot as plt

# CSV-Datei laden
df = pd.read_csv('data/null_shap_dependency_wohn_leer_mit_Y.csv')

# Scatterplot erstellen
plt.figure(figsize=(8, 6))
plt.scatter(df['wohn_leer'], df['preis_miet_best'], alpha=0.6)
plt.xlabel('wohn_leer')
plt.ylabel('preis_miet_best')
plt.title('Scatterplot: wohn_leer vs. preis_miet_best')
plt.grid(True)
plt.show()
```

```python
correlation = df['wohn_leer'].corr(df['preis_miet_best'])
print("Korrelation zwischen wohn_leer und preis_miet_best:",correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('data/testdataset.CSV', sep=';', on_bad_lines='skip')
# Nur die gewnschten Spalten auswhlen
selected_df = df[[
    'preis_miet_best', 'teilz_w', 'pendel', 'teilz_insg', 'ew_sportv', 'alq', 'teilz_m',
    'fl_suv', 'hh_veink', 'reg_energie', 'bev_dicht', 'bev_u18', 'oenv', 'bev_ausw',
    'wohn_MFH', 'GKID', 'wahl_beteil', 'heiz_wohn', 'sozsich', 'fl_wald', 'auto'
]]

# String-Kommas in Punkten umwandeln und in float konvertieren (auer 'GKID')
for col in selected_df.columns:
    if selected_df[col].dtype == object and col != 'GKID':
        selected_df[col] = selected_df[col].str.replace(',', '.').astype(float)

# Korrelationsmatrix berechnen
corr_matrix = selected_df.corr()

# Heatmap plotten
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Korrelationsmatrix Heatmap (Ausgewhlte Variablen)")
```
